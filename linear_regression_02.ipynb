```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from importlib import reload

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error
```

```python
cars = pd.read_csv("dataset/cars/cars.csv")
cars.head()
```

```python
cars.info()
```

```python
cars = cars[cars["Price in thousands"] != "."]
cars["Price in thousands"] = pd.to_numeric(cars["Price in thousands"])
cars["Horsepower"] = pd.to_numeric(cars["Horsepower"])
```

```python
plt.figure(figsize=(16,8))
plt.scatter(
    cars["Horsepower"],
    cars["Price in thousands"],
    c="black"
)
plt.xlabel = "horsepower"
plt.ylabel = "price"
```

```python
x = cars["Horsepower"].values.reshape(-1,1)
y = cars["Price in thousands"].values.reshape(-1,1)
```

```python
reg = linear_model.LinearRegression()
reg.fit(x, y)
```

```python
print(reg.coef_[0][0])
print(reg.intercept_[0])
predictions = reg.predict(x)
```

```python
print(predictions)
```

```python
plt=reload(plt)
plt.figure(figsize=(16,8))
plt.scatter(cars["Horsepower"], cars["Price in thousands"], c="black")
plt.plot(cars["Horsepower"], predictions, c="red", linewidth=2)

plt.xlabel("Horsepower")
plt.ylabel("Prices")
plt.show()
```
